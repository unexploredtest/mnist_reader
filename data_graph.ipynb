```python
from random import randint

import numpy as np
import matplotlib.pyplot as plt

from mnist_reader import read_idx_file
```

### Viewing a random number from MNIST training data





```python
# Loading data
training_data_images = read_idx_file("data/train-images-idx3-ubyte")
training_data_labels = read_idx_file("data/train-labels-idx1-ubyte")
```

```python
data_len = len(training_data_labels)
random_index = randint(0, data_len - 1)

plt.imshow(training_data_images[random_index], cmap='gray')
plt.title(f'Training data number {random_index+1}, digit is: {training_data_labels[random_index]}')
plt.show()
```

### Viewing a random number from MNIST testing data
